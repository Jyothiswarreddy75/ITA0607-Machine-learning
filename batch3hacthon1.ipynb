### Hospital Patient Diagnosis: Predicting Disease from Symptoms

This example demonstrates a basic approach to predict a disease based on a list of symptoms using a machine learning model. For a real-world application, this would require a comprehensive dataset, more advanced feature engineering, and robust model validation.

#### 1. Generate Sample Data

First, let's create a synthetic dataset of symptoms and corresponding diseases. In a real application, you would load this data from a CSV, database, or other source.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MultiLabelBinarizer

# Sample data: symptoms and diseases
data = {
    'symptoms': [
        ['fever', 'cough', 'fatigue'],
        ['fever', 'headache', 'sore throat'],
        ['cough', 'chest pain', 'shortness of breath'],
        ['headache', 'nausea', 'vomiting'],
        ['fatigue', 'muscle aches', 'joint pain'],
        ['fever', 'fatigue', 'rash'],
        ['sore throat', 'cough', 'fever'],
        ['chest pain', 'shortness of breath', 'dizziness'],
        ['nausea', 'vomiting', 'diarrhea'],
        ['joint pain', 'swelling', 'stiffness']
    ],
    'disease': [
        'Common Cold',
        'Flu',
        'Bronchitis',
        'Migraine',
        'Fibromyalgia',
        'Measles',
        'Common Cold',
        'Heart Condition',
        'Food Poisoning',
        'Arthritis'
    ]
}

df = pd.DataFrame(data)
display(df.head())
```

#### 2. Preprocess Data

Symptoms are categorical and often multi-valued. We'll use `MultiLabelBinarizer` to convert the list of symptoms into a binary matrix (one-hot encoding), where each symptom becomes a column. The diseases will be encoded using `factorize`.

```python
# Convert symptoms into a binary matrix (one-hot encoding)
mlb = MultiLabelBinarizer()
X = mlb.fit_transform(df['symptoms'])
symptom_df = pd.DataFrame(X, columns=mlb.classes_)

# Encode target variable (disease)
y, unique_diseases = pd.factorize(df['disease'])

display(symptom_df.head())
print(f"Unique Diseases: {unique_diseases.tolist()}")
```

#### 3. Train a Machine Learning Model

We'll use a `DecisionTreeClassifier` for this example. This is a simple yet effective model for classification tasks. For more complex problems, you might consider models like RandomForest, Gradient Boosting, or neural networks.

```python
# Split data into training and testing sets (optional for this small demo, but good practice)
X_train, X_test, y_train, y_test = train_test_split(symptom_df, y, test_size=0.2, random_state=42)

# Initialize and train the model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

print("Model trained successfully!")
```

#### 4. Make Predictions

Now, let's use the trained model to predict the disease for a new set of symptoms. We need to ensure the input symptoms are in the same format (one-hot encoded) as the training data.

```python
def predict_disease(symptoms_list):
    # Create a dummy DataFrame with all possible symptoms, then fill with 0s
    input_symptoms_encoded = pd.DataFrame(0, index=[0], columns=mlb.classes_)

    # Mark 1 for the present symptoms
    for symptom in symptoms_list:
        if symptom in input_symptoms_encoded.columns:
            input_symptoms_encoded[symptom] = 1

    # Make prediction
    prediction_index = model.predict(input_symptoms_encoded)[0]
    predicted_disease = unique_diseases[prediction_index]
    return predicted_disease

# Example usage:
symptoms_to_check = ['fever', 'cough', 'fatigue']
predicted_disease = predict_disease(symptoms_to_check)
print(f"Given symptoms: {symptoms_to_check}")
print(f"Most likely disease: {predicted_disease}")

symptoms_to_check_2 = ['headache', 'nausea']
predicted_disease_2 = predict_disease(symptoms_to_check_2)
print(f"\nGiven symptoms: {symptoms_to_check_2}")
print(f"Most likely disease: {predicted_disease_2}")
```

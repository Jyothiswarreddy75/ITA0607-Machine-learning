```python
#Activation Function TANH :
import numpy as np
# Input values
x = np.array([-3, -2, -1, 0, 1, 2, 3])
# Apply Tanh activation function
output = np.tanh(x)
print("Input Values:")
print(x)
print("\nTanh Output:")
print(output)
#Activation Function ReLu :
import numpy as np
# Input values
x = np.array([-5, -2, 0, 3, 7])
# Apply ReLU activation function
relu_output = np.maximum(0, x)
print("Input Values:")
print(x)
print("\nReLU Output:")
print(relu_output)
import numpy as np
marks = np.array([-10, 20, 50, -30, 80])
relu_marks = np.maximum(0, marks)
print("Original Marks :", marks)
print("After ReLU     :", relu_marks)
#Activation Function Sigmoid :
import numpy as np
# Input values
x = np.array([-5, -2, 0, 2, 5])
# Apply Sigmoid activation function
sigmoid_output = 1 / (1 + np.exp(-x))
print("Input Values:")
print(x)
print("\nSigmoid Output:")
print(sigmoid_output)
#Activation Function SOFTMAX :
import numpy as np
marks = np.array([85, 70, 60])
softmax = np.exp(marks) / np.sum(np.exp(marks))
print("Scores :", marks)
print("Softmax:", softmax)
```

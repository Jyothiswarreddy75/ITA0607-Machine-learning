```python
import pandas as pd

data = pd.read_csv("candidate.csv")

concepts = data.iloc[:, :-1].values
target = data.iloc[:, -1].values

S = concepts[0].copy()
G = [['?' for _ in range(len(S))]] # Use _ for unused loop variable

for i, h in enumerate(concepts):
    if target[i] == "Yes":
        for x in range(len(S)):
            if h[x] != S[x]:
                S[x] = '?'

    elif target[i] == "No":
        for x in range(len(S)):
            if h[x] != S[x]:
                # Note: This logic for updating G in the 'No' case is a simplification
                # and might not fully align with the standard Candidate Elimination algorithm's G-set update.
                # In a full CE algorithm, the G-set would be refined to exclude the negative instance.
                G.append(S.copy())

    print("\nSpecific Boundary:")
    print(S)

    print("\nGeneral Boundary:")
    for g_hyp in G: # Renamed 'g' to 'g_hyp' to avoid conflict with the list name 'G'
        print(g_hyp)
```
